# clean the data

```python
student = {
    'first_name': ['asmaa','esraa','mena','mona', np.nan, None, 'NA'], 
    'last_name': ['mohamed','wael','ahmed','gamal', np.nan, np.nan, 'missing'], 
    'email': ['asma@gmail.com', 'esraa@email.com', 'mena@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'missing'],
    'score':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],
    'birth_day':['2-1-2003','3-2-2004','4-4-2005','5-5-2003',None,np.nan,'missing']
}
df=pd.DataFrame(student)
df
```

```python
#show null values count
df.isna().sum()
```
